```python
df1['Embarked'].value_counts()
```

```python
df1['Sex'].value_counts()
```

```python
sex = pd.get_dummies(df1['Sex'],drop_first=True)
```

```python
sex
```

```python
Embark = pd.get_dummies(df1['Embarked'],drop_first=True)
Embark
```

```python
pcl = pd.get_dummies(df1['Pclass'],drop_first=True)
pcl

```

```python
df1 = pd.concat([df1,sex,Embark,pcl],axis=1)
```

```python
df1
```

```python
df1.drop(['Sex','Embarked','PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
df1
```

```python
df1.drop('Pclass',axis=1,inplace=True)
```

```python
df1
```

```python
x=df1.drop('Survived',axis=1)
y=df1['Survived']
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
x_train ,x_test,y_train ,y_test =train_test_split(x,y,test_size=0.3)
```

```python
y_test
```

```python
log_model = LogisticRegression()
```

```python
log_model.fit(x_train,y_train)
```

```python
predict = log_model.predict(x_test)
```

```python
print(classification_report(y_test,predict))
```

```python
confusion_matrix(y_test,predict)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predict)
```
